```python
data = token_transfers_df.copy()
# data.to_excel("transactions.xlsx")
data[["tokenDecimal", "value"]] = data[["tokenDecimal", "value"]].astype(float)

valid_tokens = data[data['hash'].isin(token_transfers_all['hash'])]['tokenSymbol'].unique().tolist()
data = data[data['tokenSymbol'].isin(valid_tokens)]

data  #.to_dict('records')
```

```python
# Define wallet address
wallet_address = addresses[wallet_name]

# Initialize an empty dictionary to store token balances
balances = {}

# Function to update the balance of tokens based on inflow and outflow
def update_balance(row):
    token = row['tokenSymbol']
    decimals = row['tokenDecimal']
    value = row['value'] / (10 ** decimals)  # Adjust for decimals
    
    # Initialize balance for the token if it doesn't exist
    if token not in balances:
        balances[token] = 0

    # Check if it's an inflow or outflow
    if row['to'].lower() == wallet_address.lower():
        balances[token] += value  # Inflow
    elif row['from'].lower() == wallet_address.lower():
        balances[token] -= value  # Outflow

# Apply the function to each row in the DataFrame
data.apply(update_balance, axis=1)

# Print the final balances
for token, balance in balances.items():
     if balance != 0 and abs(balance) >= 0.0001:
        print(f"Token: {token}, Balance: {balance}")
```
